```python
import csv
import numpy as np
import pandas as pd
import collections
import re
from datetime import datetime
#datetime_object = datetime.strptime(data_time[0], '%Y-%m-%d %H:%M:%S')

file = open("reddit_database.csv")

reader = csv.reader(file)
```

```python
rows = []
for row in reader:
    rows.append(row)
```

```python
data = []
data_text = []
data_time = []
d1 = datetime(2011, 1, 3)
for row in rows[1:]:
    time1 = datetime.strptime(row[0], '%Y-%m-%d %H:%M:%S').date()
    if time1<d1.date():
        data.append([row[0], row[3]+' '+ row[12]])
        data_text.append(row[3]+' '+ row[12])
        data_time.append(time1)
```

```python
for i in range(len(data_text)):
    data_text[i] = re.sub(r"[^a-zA-Z]", " ", data_text[i].lower())
```

```python
import gensim
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary
```

```python
a = []
for i in range(len(data_text)):
    a.append(data_text[i].split())
a = a[:]
```

```python
a = [[word for word in doc if word not in stop_words] for doc in a]
```

```python
print(a)
```

```python
import gensim.corpora as corpora
id1 = corpora.Dictionary(a)
corpus = [id1.doc2bow(text) for text in a]
```

```python
num_topics = 5
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id1,
                                       num_topics=num_topics)
```

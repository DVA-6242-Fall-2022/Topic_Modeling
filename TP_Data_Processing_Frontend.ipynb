```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
# name of the subreddit to process
subreddit = 'movies'
```

```python
time_data = pd.read_csv(f'../Desktop/DVA Project Data/{subreddit.capitalize()}/topics_over_time_{subreddit}.csv')
```

```python
time_data = time_data.drop(time_data.index[time_data['Topic'] == -1], axis=0)
```

```python
time_data
```

```python
sentiment_data = pd.read_csv(f'../Desktop/DVA Project Data/{subreddit.capitalize()}/{subreddit}_comment_sentiment.csv')
```

```python
sentiment_data = sentiment_data.drop(sentiment_data.index[sentiment_data['topic'] == -1], axis=0)
```

```python
sentiment_data
```

```python
timestamp_to_month = lambda x: datetime.strptime(x, '%Y-%m-%d').strftime('%B %Y')
```

```python
string_split_date = lambda x: x.split(' ')[0]
```

```python
time_data['Timestamp'] = time_data['Timestamp'].apply(string_split_date)
```

```python
time_data.head(5)
```

```python
time_data = time_data.replace('2019-12-31', '2020-01-01')
```

```python
time_data['Timestamp'] = time_data['Timestamp'].apply(timestamp_to_month)
```

```python
time_data
```

```python
np.unique(time_data['Timestamp'])
```

```python
sentiment_data['created_dt_month'] = sentiment_data['created_dt_month'].apply(timestamp_to_month)
```

```python
np.unique(sentiment_data['created_dt_month'])
```

```python
final_df = time_data.merge(sentiment_data, how='inner', left_on=['Topic', 'Timestamp'], right_on=['topic', 'created_dt_month'])
```

```python
final_df
```

```python
output_df = final_df[['Topic', 'Words', 'Frequency', 'Timestamp', 'comment_sentiment', 'name']]
```

```python
output_df.to_csv(f'../Desktop/DVA Project Data/{subreddit.capitalize()}/topics_sentiments_{subreddit}.csv', index=False)
```
